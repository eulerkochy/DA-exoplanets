All of us know this poem by Jane Taylor. How exactly, do we locate these diamonds in the sky?

Star position is the apparent location of any given star in the sky, which seems fixed onto an arbitrary sphere centered on Earth. The location is defined by a pair of angular coordinates relative to the celestial equator: right ascension (α) and declination (δ). This pair based the equatorial coordinate system.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

df=pd.read_csv(filepath_or_buffer='Documents/oec.csv')
sns.set()
```

```python
features=['PlanetIdentifier','RightAscension', 'Declination']
star=df[features].dropna()
def func(x):
    y=x.strip().split()
    return float( int(y[0])+ int(y[1])/60 + float(y[2])/3600)
star['RightAscensionRad']=star['RightAscension'].map(func) 
star['DeclinationRad']=star['Declination'].map(func)
star.plot.scatter(x='DeclinationRad',y='RightAscensionRad')
```

```python
sns.jointplot(x=['DeclinationRad'], y=['RightAscensionRad'], 
              data=star, kind='scatter')
```

```python
sns.jointplot('DeclinationRad', 'RightAscensionRad', data=star,
              kind="kde", color="r");
```
