```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

df=pd.read_csv(filepath_or_buffer='Documents/oec.csv')
```

# On DiscoveryMethod 

It will help us in analysing recent trends in discovering exoplanets!



```python
df['DiscoveryMethod'].value_counts()
```

```python
df['DiscoveryMethod'].value_counts().plot.pie()
plt.title('No. of planets discovered')
```


  
This shows transit is the most popular method of discovery.
  



```python
sns.distplot(df[df['DiscoveryMethod']=="transit"]['DiscoveryYear'].dropna())
```

```python
sns.distplot(df[df['DiscoveryMethod']=="RV"]['DiscoveryYear'].dropna())
```

```python
sns.distplot(df[df['DiscoveryMethod']=="imaging"]['DiscoveryYear'].dropna())
```

```python
sns.distplot(df[df['DiscoveryMethod']=="microlensing"]['DiscoveryYear'].dropna())
```

```python
sns.distplot(df[df['DiscoveryMethod']=="timing"]['DiscoveryYear'].dropna())
```

Combining all of them to be displayed in a single plot

```python
features=['DiscoveryYear','DiscoveryMethod']
cmb=df.groupby(features)[[features[1]]].agg(['count'])
cmb.unstack().plot(kind='bar', stacked=True)
plt.title('Discovery Methods Year-Wise')
```
